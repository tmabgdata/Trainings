```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tmabgdata/data_analysis/master/pokemon_analysis/pokemon_complete_data_processing.csv')
```

```python
df.head(5)
```

```python
# On average, which type of pokemon has more total power, compared to other types?
# Em média, qual tipo de pokémon tem mais poder total, comparado aos outros tipos?
# => 1º Dragon | 2º Steel | 3º Flying
df.groupby(['Type 1']).mean().sort_values('Total Pw', ascending = False)
```

```python
attribute = 'Attack'

bins = range(0, 201, 20)
plt.xticks(bins)

plt.title('Pokemon distribution by ' + attribute)
plt.xlabel('Attribute value ' + attribute)
plt.ylabel('Amount of pokemon')
plt.hist(df[attribute], bins = bins)
```

```python
attribute = 'Attack'

types = df.groupby('Type 1').mean().sort_values(attribute, ascending = False)
line = types.index.tolist()
column = types[attribute].tolist()
figure = plt.figure(figsize=(14,4)).add_axes([0, 0, 1, 1])
figure.bar(line, column)

plt.xlabel('Pokemons Types')
plt.ylabel('Average Attribute Value ' + attribute)
plt.title('Average Attribute Value ' + attribute + ' Pokemon Type')

for i, v in enumerate(column):
  plt.annotate(f'{v:.2f}', (i-0.3, v+1))


plt.show()
```
